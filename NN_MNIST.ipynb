```python
import sys, numpy as np
```

```python
from keras.datasets import mnist
```

```python
(x_train,y_train),(x_test,y_test)=mnist.load_data()
```

```python
print(x_train.shape)
```

```python
x_train,y_train=(x_train[0:1000].reshape(1000,28*28)/255, y_train[0:1000])
```

```python
one_hot_labels=np.zeros((len(y_train),10))

for i,l in enumerate(y_train):
  one_hot_labels[i][l]=1

y_train=one_hot_labels
```

```python
x_test=x_test.reshape(len(x_test),28*28)/255

y_hot_labels=np.zeros((len(y_test),10))

for i,l in enumerate(y_test):
  y_hot_labels[i][l]=1

y_test=y_hot_labels
```

```python
def relu(x):
  return x*(x>0)

def relu_derivative(x):
  return x>0
```

```python
alpha, iterations, hidden_size, pixels_per_image, num_labels = (0.005, 350, 40, 784, 10)
```

```python
weights_0_1=0.2*np.random.random((pixels_per_image,hidden_size))-0.1
weights_1_2=0.2*np.random.random((hidden_size,num_labels))
```

```python
error_all=0

for iter in iterations:
  error_all=0
  for i in range(len(x_train)):
    ip=x_train[i:i+1]
    op=y_train[i:i+1]

    layer_1=relu(np.dot(ip,weights_0_1))
    layer_2=np.dot(layer_1,weights_1_2)

    layer_2_error=np.sum((layer_2-op)**2)
    layer_2_delta=(layer_2-op)

    layer_1_delta=np.dot(layer_2_delta,weights_1_2.T))*relu_derivative(layer_1)

    weights_1_2-=alpha*np.dot(layer_1.T,layer_2_delta)
    weights_0_1-=alpha*np.dot(ip.T,layer_1_delta)

    error_all+=layer_2_error

    if iter%10==0:
      print("error at iter: ",error_all)


```
